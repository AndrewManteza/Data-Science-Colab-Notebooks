```
import pandas as pd
import numpy as np
import math

def split_data_by_class(data,train_percent,test_percent):

  cols = data.columns

  distinct_class_val = data[cols[-1]].unique()

  data_dic= {}

  index_start = 0

  for i in distinct_class_val:

    datas = data[data[cols[-1]] == distinct_class_val[index_start]].values.tolist()

    total_rows = len(datas)


    train_no = round(total_rows*(train_percent/100))
    test_no =  round(total_rows*(test_percent/100))
    data_dic.update({i:{'train':datas[:train_no] ,'test':datas[train_no:]}})

    index_start+=1

  merge_train = data_dic[distinct_class_val[0]]['train']

  merge_test = data_dic[distinct_class_val[0]]['test']


  for i in range(len(distinct_class_val)-1):

    merge_train = np.concatenate((merge_train, data_dic[distinct_class_val[i+1]]['train'] ) )

    merge_test =  np.concatenate((merge_test, data_dic[distinct_class_val[i+1]]['test']) )


  return pd.DataFrame(merge_train,columns = cols), merge_test


def get_mean_std(data):


  cols = data.columns

  distinct_class_val = data[cols[-1]].unique()

  index_start = 0

  datas = {}

  total_rows = 0

  for i in distinct_class_val:

    awits = data[data[cols[-1]] == distinct_class_val[index_start]]

    awits = awits[awits.columns[:-1]].astype(float)

    stds = awits.std()

    total_rows += len(awits)

    means = awits.mean()

    datas[i] = {}

    for j in range(len(cols)-1):

      datas[i].update({ cols[j] : {'mean': means[cols[j]],'std': stds[cols[j]] } })


    datas[i].update({'total_rows': len(awits)})

    index_start+=1

  datas.update({'cols':cols[:-1]})

  datas.update({'total_rows':total_rows})

  return datas
```

```
data = pd.read_csv('/content/Iris.csv - Iris.csv.csv')

data, x  = split_data_by_class(data,80,20)

data_stat = get_mean_std(data)

classes = list(data_stat.keys())[:-2]

for i in range(len(x)):

  list_vals = []

  for j in classes:

    prob = 1

    index_input = 0

    for k in data_stat['cols']:

      expo = math.e**(-1*( (float(x[i][index_input])-data_stat[j][k]['mean'])**2 / ( 2*(data_stat[j][k]['std'])**2  )) )

      prob *= (1/(data_stat[j][k]['std'] * math.sqrt(2*math.pi) ) )*expo

      index_input+=1

    prob *= data_stat[j]['total_rows']/data_stat['total_rows']

    list_vals.append((j,prob))

  list_vals.sort(key=lambda tup: tup[1] )


  print(x[i],i+1,':',list_vals[-1][0])
```
